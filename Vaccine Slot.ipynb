```python
with open('district_ids.json') as f:
    dist_ids = json.load(f)
```

```python
state = "Gujarat"
district = "Kutch"
date = "05-05-2021"
pin = "370110"
age = 18 #1 : all, 18: 18-45 , 
dist_id = dist_ids[state][district]
print(dist_id)

url_dist = f"https://cdn-api.co-vin.in/api/v2/appointment/sessions/public/calendarByDistrict?district_id={dist_id}&date={date}"
url_pin = f"https://cdn-api.co-vin.in/api/v2/appointment/sessions/public/calendarByPin?pincode={pin}&date={date}"

dist = 1
```

```python
if dist:
    url = url_dist
else:
    url = url_pin

#request
resp = eval(requests.get(url, proxies=urllib.request.getproxies()).text)
```

```python
all_centers = resp['centers']
all_centers[1]
```

```python
min_age = 18 if age == 18 else 45
print(min_age)
available = {}
for each in all_centers:
    center_name = each["name"].strip()
    if each['sessions']: 
        for sess in each['sessions']:
            
            if sess['min_age_limit'] == min_age and sess["available_capacity"] > 0:
                data = {"available_capacity":sess["available_capacity"]
                                             ,"date":sess["date"]}
                if center_name not in available:
                    available[center_name] = [data]
                else:
                    available[center_name].append(data)
                print("available!")
   
        
        
```
